```python
import cv2
import numpy as np
import random
```

```python
w = 400
h = 400

def rnd(start, end):
    return random.randrange(start, end)

def draw_ellipse(img, angle):

    cv2.ellipse(img, 
                (rnd(0, 400), rnd(0, 400)), 
                (rnd(10, 200), rnd(10, 200)),
                angle,
                0, 
                360, 
                (rnd(0, 255), rnd(0, 255), rnd(0, 255)), 
                -1, 
                lineType=cv2.LINE_AA)

def draw_triangle(image):

    pt1 = (rnd(0, 400), rnd(0, 400))
    pt2 = (rnd(0, 400), rnd(0, 400))
    pt3 = (rnd(0, 400), rnd(0, 400))

    triangle_points = np.array([pt1, pt2, pt3])

    cv2.drawContours(image, 
                    [triangle_points], 
                    0, 
                    (rnd(0, 255), 
                    rnd(0, 255), 
                    rnd(0, 255)), 
                    -1, 
                    lineType=cv2.LINE_AA)

def draw_rectangle(image):

    pt1 = (rnd(0, 400), rnd(0, 400))
    pt2 = (rnd(0, 400), rnd(0, 400))
    pt3 = (rnd(0, 400), rnd(0, 400))
    pt4 = (rnd(0, 400), rnd(0, 400))

    rectangle_points = np.array([pt1, pt2, pt3, pt4])

    cv2.drawContours(image, 
                     [rectangle_points], 
                     0, 
                     (rnd(0, 255), rnd(0, 255), rnd(0, 255)), 
                     -1, 
                     lineType=cv2.LINE_AA)

def draw_star(image):
    
    pt1 = (rnd(0, 400), rnd(0, 400))
    pt2 = (rnd(0, 400), rnd(0, 400))
    pt3 = (rnd(0, 400), rnd(0, 400))
    pt4 = (rnd(0, 400), rnd(0, 400))
    pt5 = (rnd(0, 400), rnd(0, 400))
    
    color = (rnd(0, 255), rnd(0, 255), rnd(0, 255))
    centroid =((pt1[0] + pt2[0] + pt3[0] + pt4[0] + pt5[0]) // 5, (pt1[1] + pt2[1] + pt3[1] + pt4[1] + pt5[1]) // 5)

    cv2.line(image, pt1, centroid, color, 2, lineType=cv2.LINE_AA)
    cv2.line(image, pt2, centroid, color, 2, lineType=cv2.LINE_AA)
    cv2.line(image, pt3, centroid, color, 2, lineType=cv2.LINE_AA)
    cv2.line(image, pt4, centroid, color, 2, lineType=cv2.LINE_AA)
    cv2.line(image, pt5, centroid, color, 2, lineType=cv2.LINE_AA)

image = np.ones((400, 400, 3), dtype=np.uint8) * np.array([0, 0, 255], np.uint8)
window = "Shapes"
```

```python
for i in range(5):
    draw_ellipse(image, 0)
    draw_triangle(image)
    draw_rectangle(image)
    draw_star(image)

output_image = image

    
```

```python
#Additive White Gaussian Noise

gaussian_noise = np.random.normal(0, 0.1, output_image.shape).astype('float32')
image_with_noise = output_image.astype('float32') / 255.0 + gaussian_noise
image_with_noise = np.clip(image_with_noise, 0, 1)
image_with_noise = (image_with_noise * 255.0).astype('uint8')

cv2.imshow(window, image_with_noise)
cv2.moveWindow(window, 0, 200)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
